```python
churn_df = df.copy()
```

```python
churn_df['Frequency_Risk'] = churn_df['Previous Purchases'].apply(
    lambda x: 1 if x < churn_df['Previous Purchases'].median() else 0
)
```

```python
churn_df['Subscription_Risk'] = churn_df['Subscription Status'].apply(
    lambda x: 1 if x == 'No' else 0
)
```

```python
churn_df['Rating_Risk'] = churn_df['Review Rating'].apply(
    lambda x: 1 if x < 3.5 else 0
)
```

```python
churn_df['Churn_Risk_Score'] = (
    churn_df['Frequency_Risk'] +
    churn_df['Subscription_Risk'] +
    churn_df['Rating_Risk']
)
```

```python
def churn_category(score):

    if score >= 2:
        return "High Risk"

    elif score == 1:
        return "Medium Risk"

    else:
        return "Low Risk"


churn_df['Churn_Category'] = churn_df['Churn_Risk_Score'].apply(
    churn_category
)
```

```python
churn_df.groupby(
    ['CLV_Segment','Churn_Category']
).size()
```

```python
revenue_at_risk = churn_df[
    churn_df['Churn_Category']=="High Risk"
].groupby(
    'CLV_Segment'
)['Estimated_CLV'].sum()

revenue_at_risk
```

```python
product_perf = df.groupby('Item Purchased').agg({
    'Purchase Amount (USD)': 'sum',
    'Review Rating': 'mean'
}).reset_index()

# thresholds
sales_median = product_perf['Purchase Amount (USD)'].median()
rating_median = product_perf['Review Rating'].median()

def classify(row):
    if row['Purchase Amount (USD)'] >= sales_median and row['Review Rating'] >= rating_median:
        return "Hero Product"
    elif row['Purchase Amount (USD)'] >= sales_median:
        return "Problem Product"
    elif row['Review Rating'] >= rating_median:
        return "Hidden Gem"
    else:
        return "Low Performer"

product_perf['Product_Type'] = product_perf.apply(classify, axis=1)

```

```python
persona_df = churn_df.copy()
```

```python
def assign_persona(row):

    if (
        row['CLV_Segment'] == 'VIP'
        and row['Subscription Status'] == 'Yes'
    ):
        return "VIP Loyalist"

    elif (
        row['Discount Applied'] == 'Yes'
    ):
        return "Deal Seeker"

    elif (
        row['Churn_Category'] == 'High Risk'
    ):
        return "At-Risk Customer"

    else:
        return "Regular Shopper"


persona_df['Customer_Persona'] = persona_df.apply(
    assign_persona,
    axis=1
)
```

```python
persona_df['Customer_Persona'].value_counts()
```

```python
persona_summary = persona_df.groupby(
    'Customer_Persona'
).agg(
    Customers=('Customer ID','count'),
    Avg_CLV=('Estimated_CLV','mean'),
    Avg_Purchase=('Purchase Amount (USD)','mean')
)

persona_summary
```
